```python
with open('sample2.txt') as f:
  text = str(f.read())
```

```python
print(text)
```

```python
text = text.replace('\n', ' ')
text = text.replace('-', ' ')
text = text.split(' ')
```

```python
new_d = [] 
for list_2 in text:
  
  list_2 = list_2.replace(',','')
  list_2 = list_2.replace('.','')
  list_2 = list_2.replace(':','')
  list_2 = list_2.replace(';','')
  list_2 = list_2.replace('?','')
  list_2 = list_2.replace('!','')
  list_2 = list_2.lower()
  new_d.append(list_2)
        #print(list_2)
```

```python
import mr4mp

from random import choice
from string import ascii_lowercase
```

```python
def map_1(key):
    return {key:1}

def reduce_1(pairs1,pairs2):
    #print("pairs1 keys:", list(pairs1.keys()))
    #print("pairs2 keys:", list(pairs2.keys()))
    keys = pairs1.keys() | pairs2.keys()
    merged = {k: (pairs1.get(k, 0) + pairs2.get(k, 0)) for k in keys}
    return merged
```

```python
import multiprocessing
multiprocessing.set_start_method("fork", force=True)
```

About 4 years ago, when this code was done I were using Python 3.7 or earlier. In those versions, macOS used a method called fork to start new processes.
-fork literally "clones" the memory of your Jupyter Notebook
-spawn (the new default since Python 3.8 on Mac) starts a brand new, empty Python process that has to "import" your code. Since Jupyter cells aren't a real file, the import fails.

that is why the previous 2 lines of code were added

```python
from timeit import default_timer


if __name__ == "__main__":

    start = default_timer()
    pool = mr4mp.pool()
    #help(pool.mapreduce)
    out = pool.mapreduce(map_1, reduce_1, new_d)
    pool.close()
    print("Finished in " + str(default_timer()-start) + "s using " + str(len(pool)) + " process(es).")
```
